```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker()
np.random.seed(42)
random.seed(42)
```

```python
# Define possible values for categorical columns
nationalities = ['Australian', 'New Zealander', 'British', 'Indian']
occupations = ['Engineer', 'Teacher', 'Banker', 'Doctor', 'Artist']
fee_structures = ['Standard', 'Premium', 'Basic']
loyalty_classes = ['Gold', 'Silver', 'Platinum', 'Bronze']
risk_weights = ['Low', 'Medium', 'High']
genders = [1, 2]  # Assuming 1 = Male, 2 = Female
```

```python
# Generate synthetic data
rows = []
for i in range(200):
    row = {
        "Client ID": f"C{i+1000}",
        "Name": fake.name(),
        "Age": random.randint(22, 65),
        "Location ID": random.randint(100, 999),
        "Joined Bank": fake.date_between(start_date='-10y', end_date='today').strftime('%d-%m-%Y'),
        "Banking Contact": fake.phone_number(),
        "Nationality": random.choice(nationalities),
        "Occupation": random.choice(occupations),
        "Fee Structure": random.choice(fee_structures),
        "Loyalty Classification": random.choice(loyalty_classes),
        "Estimated Income": round(random.uniform(30000, 500000), 2),
        "Superannuation Savings": round(random.uniform(10000, 300000), 2),
        "Amount of Credit Cards": random.randint(0, 5),
        "Credit Card Balance": round(random.uniform(0, 50000), 2),
        "Bank Loans": round(random.uniform(0, 200000), 2),
        "Bank Deposits": round(random.uniform(1000, 150000), 2),
        "Checking Accounts": round(random.uniform(0, 50000), 2),
        "Saving Accounts": round(random.uniform(0, 80000), 2),
        "Foreign Currency Account": round(random.uniform(0, 30000), 2),
        "Business Lending": round(random.uniform(0, 100000), 2),
        "Properties Owned": random.randint(0, 5),
        "Risk Weighting": random.choice(risk_weights),
        "BRId": random.randint(1, 5),
        "GenderId": random.choice(genders),
        "IAId": random.randint(100, 999)
    }
    rows.append(row)
```

```python
df_synthetic = pd.DataFrame(rows)
df_synthetic.head()
```

```python
df_synthetic.to_csv("sample_data.csv", index=False)
```
